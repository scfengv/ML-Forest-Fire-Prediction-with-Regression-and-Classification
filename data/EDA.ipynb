```python
import pylab
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")
%config InlineBackend.figure_format = 'retina'
```

```python
f = '/Users/shenchingfeng/GitHub/ML-Forest-Fire-Prediction-with-Regression-and-Classification/data/forestfires.csv'
df = pd.read_csv(f)
d = df.copy()
d_n0 = d[d['area'] > 0].reset_index(drop = True)
```

# Exploratory data analysis

```python
plt.figure(figsize = (10, 6))

df['area'].hist(bins = 100)
plt.title('Distribution of forest fire area')
plt.xlabel('forest fire area')
plt.ylabel('numbers')
plt.show()
```

```python
plt.figure(figsize = (10, 6))

plt.scatter(df['X'], df['Y'], s = 10 * np.log1p(df['area']), color = 'red', alpha = 0.2, label = 'Burnt area')

plt.title('Burnt area in different regions', fontsize = 20)

plt.xlabel('X', fontsize = 16, fontweight='bold')
plt.ylabel('Y', fontsize = 16, fontweight='bold')

plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

plt.legend(fontsize = 12)

plt.show()
```

```python
month_map = {
    "jan": 1, "feb": 2, "mar": 3, "apr": 4, "may": 5, "jun": 6, "jul": 7, "aug": 8, "sep": 9, "oct": 10, "nov": 11, "dec": 12
}

df['month'] = df['month'].map(month_map)
fire_month = df['month'].value_counts().sort_index()
plt.bar(x = fire_month.index, height = fire_month.values)

plt.title("Forest Fire by month")
plt.xlabel("Month")
plt.ylabel("Forest Fire")
plt.show()
```

```python
day_map = {
    "mon": 1, "tue": 2, "wed": 3, "thu": 4, "fri": 5, "sat": 6, "sun": 7
}

df['day'] = df['day'].map(day_map)
fire_day = df['day'].value_counts().sort_index()
plt.bar(x = fire_day.index, height = fire_day.values)

plt.title("Forest Fire by Weekday")
plt.xlabel("Day")
plt.ylabel("Forest Fire")
plt.show()
```
